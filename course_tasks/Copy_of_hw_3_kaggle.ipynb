```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
encoded_data = dropped_data.copy()

encoded_data.HasChild = dropped_data.HasChild.replace(['Yes', 'No'], [1, 0])
encoded_data.HasPartner = dropped_data.HasPartner.replace(['Yes', 'No'], [1, 0])
encoded_data.IsBillingPaperless = dropped_data.IsBillingPaperless.replace(['Yes', 'No'], [1, 0])
```

```python
encoding_features = list(new_cat_cols)
# # rev_arr = ['IsSeniorCitizen', 'HasPartner', 'HasChild', 'IsBillingPaperless']
# for el in rev_arr:
#   encoding_features.remove(el)
# print(encoding_features)
```

```python
one_hot = OneHotEncoder(sparse_output=False, drop='first')

encoded_features = one_hot.fit_transform(dropped_data[encoding_features])

one_hot_df = pd.DataFrame(encoded_features, columns = one_hot.get_feature_names_out(encoding_features), index = dropped_data.index)
encoded_data = pd.concat([dropped_data, one_hot_df], axis = 1)
encoded_data.drop(columns=new_cat_cols, inplace=True)
```

```python
plt.figure(figsize=(11, 11))
corrmat = encoded_data.corr()
# dataf = corrmat.nlargest(16, 'Churn')
dataf = corrmat
cols = dataf['Churn'].index
cm = np.corrcoef(encoded_data[cols].values.T)
sns.set(font_scale = 1.25)
htmap = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, shuffle=True)
```

```python
x_train
```

```python
scaler = StandardScaler()
# scaler = RobustScaler()
scaler.fit(x_train)

x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
logistic_model = LogisticRegression()

grid_params={
    'C': [100, 10, 1, 0.1, 0.01, 0.001],
    'solver' : ['newton-cg', 'liblinear', 'sag', 'saga'],
    'penalty' : ['l1', 'l2', 'elasticnet', 'none']
}

grid_search_logistic = GridSearchCV(logistic_model, grid_params, cv = 5, refit=True)

grid_search_logistic.fit(x_train, y_train)
best_model = grid_search_logistic.best_estimator_

# Я хотел здесь написать функцию для автоматического печатания и делания всего этого барахла
```

```python
# Here's neural network

import torch.nn as nn

class Net(nn.Module):
  def __init__(self):
    super(Net, self).__init__()

    self.fc1 = nn.Linear(15, 40)
    self.act1 = nn.Sigmoid()

    self.fc2 = nn.Linear(40, 1)
    self.act2 = nn.Sigmoid()

  def forward(self, x):
    x = self.fc1(x)
    x = self.act1(x)

    x = self.fc2(x)
    x = self.act2(x)

    return x
```

```python
from tqdm.auto import tqdm
from torch.utils.data import DataLoader, TensorDataset

def train(model, loader, criterion, optimizer, num_epoch, epoch_loss):
  '''
    args:
        model - модель нейронной сети
        X и y - обучающая выборка
        criterion - функция потерь, взятая из модуля `torch.nn`
        optimizer - оптимизатор, взятый из модуля `torch.optim` // it's just stohastic gradient stepping
        num_epoch - количество эпох обучения.
    '''

  for t in tqdm(range(num_epoch)):
    arr = []
        # цикл по батчам даталоадера
    for X_batch, y_batch in loader:
            # Вычислим предсказания нашей модели
      y_pred = model(X_batch)

            # Посчитаем значение функции потерь  на полученном предсказании
      loss = criterion(y_pred, y_batch)
      arr.append(loss.item())

            # Выполним подсчёт новых градиентов
      loss.backward()
            # Выполним шаг градиентного спуска
      optimizer.step()
            # Обнулим сохраненные у оптимизатора значения градиентов
            # перед следующим шагом обучения
      optimizer.zero_grad()

      # print("curr_loss", np.mean(epoch_loss))
    epoch_loss.append(np.mean(arr))
  return model
```
