```python
# Here's neural network

import torch.nn as nn

class Net(nn.Module):
  def __init__(self):
    super(Net, self).__init__()

    self.fc1 = nn.Linear(15, 40)
    self.act1 = nn.Sigmoid()

    self.fc2 = nn.Linear(40, 1)
    self.act2 = nn.Sigmoid()

  def forward(self, x):
    x = self.fc1(x)
    x = self.act1(x)

    x = self.fc2(x)
    x = self.act2(x)

    return x
```

```python
from tqdm.auto import tqdm
from torch.utils.data import DataLoader, TensorDataset

def train(model, loader, criterion, optimizer, num_epoch, epoch_loss):
  '''
    args:
        model -   
        X  y -  
        criterion -  ,    `torch.nn`
        optimizer - ,    `torch.optim` // it's just stohastic gradient stepping
        num_epoch -   .
    '''

  for t in tqdm(range(num_epoch)):
    arr = []
        #    
    for X_batch, y_batch in loader:
            #    
      y_pred = model(X_batch)

            #        
      loss = criterion(y_pred, y_batch)
      arr.append(loss.item())

            #    
      loss.backward()
            #    
      optimizer.step()
            #      
            #    
      optimizer.zero_grad()

      # print("curr_loss", np.mean(epoch_loss))
    epoch_loss.append(np.mean(arr))
  return model
```
